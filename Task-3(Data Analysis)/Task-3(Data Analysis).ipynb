### Data Exploratory Analysis Using the SuperStore Dataset 

```r
#imprting the libraries
library(ggplot2)
```

```r
store<-read.csv("SampleSuperstore.csv")
```

```r
head(store)
```

```r
colSums(is.na(store))
```

```r
summary(store)
```

```r
str(store)
```

```r
ggplot(store)+
geom_bar(aes(x=Ship.Mode,fill=Profit))
```

### As we can see the graph the profit is lower in people using the Same day Ship Mode

```r
ggplot(store)+
geom_bar(aes(x=Segment,fill=Profit))
```

### The graph makes it clear that the Home office needs more attention as it has the lowest profit margin

```r
ggplot(store)+
geom_bar(aes(x=State,fill=Profit))
```

### This graph has large no. of variables so it gets very hard to find a perfect analysis so i think it would be better to make a graph using the regions

```r
ggplot(store)+
geom_bar(aes(x=Region,fill=Profit))
```

### Now this makes it clear the States lying in the south region has low Profit margin

```r
ggplot(store)+
geom_bar(aes(x=Sub.Category,fill=Profit))+facet_wrap(~Category)
```

### We can make a graph by enveloping the sub category into three parts which are marked by Category. But the graph doesnt give us a clear relationship between the category and sub category

```r
ggplot(store)+
geom_bar(aes(x=Quantity,fill=Profit))
```

```r
ggplot(store)+
geom_bar(aes(x=Quantity,fill=Profit))+facet_wrap(~Sub.Category)
```

### We can see that by enveloping the quantity into sub category we get a clear relationship we can say that as the quantity increaeses the profit margin decreases
